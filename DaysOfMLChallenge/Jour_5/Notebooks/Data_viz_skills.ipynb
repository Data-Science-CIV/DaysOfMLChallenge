### Visualisation de données : Quoi? Pourquoi? Comment
**Quoi ?**

> Visualiser les données se rapporte au fait de représenter un ensemble de données (en général tabulaires...mais pas toujours) à l'aide d'éléments visuels tels que __les graphiques et les cartes__, dans le but de mieux comprendre les tendances et de faire ressortir des informations qui auraient été cachées sous la forme tabulaire du jeu de données.
Cette étape se retrouve dans une partie appelée __Analyse de données exploratoire (EDA en anglais)__ qui est  __très importante__ dans le cycle d'un projet de data science car c'est la toute première partie du travail. Dans cette partie il est question de faire du __data mining__ c'est-à-dire de la __fouille de données__ pour en ressortir des information pertinentes qui vous aideront à mieux appréhender le problème à résoudre et ce, avec nettement plus d'aisance.


**Pourquoi ?**

> Avec l'avènement du __Big Data__, les données à analyser sont devenues beaucoup plus nombreuses et complexes. Il est parfois même banal de se retrouver à analyser un tableau de données contenant des centaines de lignes et colonnes, puisque dans le métier du data scientist, celui-ci sera amené à manipuler des milliers, voire des millions d'enregistrements et d'en faire ressortir des informations pertinentes exploitables par l'entité qui met à disposition ces données. 

> Si l'on suppose que vous avez en votre possession un tableau de données  __INDIVIDUSxVARIABLES quantitatives__ avec __1000 lignes et 30 colonnes__ (soit ~30000 valeurs à analyser car il est possible d'avoir des cases vides), il est bien plus aisé de voir les relations entre ces données à travers une représentation graphique plutôt que de s'obstiner à vouloir faire de la magie en regardant très fort cette matrice gigantesque. 
Avec une visualisation bien réalisée, il est possible d'avoir des intuitions sur le/les _modèle(s)_ à construire pour résoudre la tâche données.

> La visualisation permet de communiquer les hypothèses à des "non data scientists" et faciliter la compréhensions de ces hypothèses. C'est un puissant outils d'aide à la décision qui ne doit pas être baclé ou négligé.


**Comment ?**

> Visualiser un jeu de données se fait à l'aide d'outils spéciaux entre autres des logiciels et applications dédiés à cet effet. Toutefois, il existe de nombreux modules utilisables à travers les langages de programmation, qui permettent de réaliser cette visualisation. Par la suite nous allons en voir certains.

### Diagramme en bâton
> Diagramme dans lequel on représente les valeurs d'une distribution d'une variable statistique quantitative discrète à l'aide de segments verticaux ou horizontaux. Le diagramme à bâtons s'apparente à un diagramme à bandes.

source : www.lexique.netmath.ca 

```python
# voici un histogramme qui est nettement plus informatif
plt.bar(x=planets.year, height=planets.mass)
plt.xlabel('année')
plt.ylabel('Masse')
plt.title("Evolution de la masse en fonction de l'année")
plt.show()
```

### Nuage de points

> Un nuage de points ou diagramme de dispersion est une représentation graphique dans un repère du plan d'une série statistique à deux variables X et Y. Chaque individu i est représenté par un point dont les coordonnées sont les valeurs respectives des variables X et Y prises par l'individu i.

source : [fr.khanacademy.org ](https://www.google.com/url?sa=t&rct=j&q=&esrc=s&source=web&cd=&cad=rja&uact=8&ved=2ahUKEwjz0MWc47bqAhUoSRUIHfrMDpcQFjABegQIChAD&url=https%3A%2F%2Ffr.khanacademy.org%2Fmath%2Fstatistics-probability%2Fdescribing-relationships-quantitative-data%2Fintroduction-to-scatterplots%2Fa%2Fscatterplots-and-correlation-review&usg=AOvVaw1Eh5UcBLJb_fHhU0qeCHXe)

```python
plt.title('enregistrement des masses par année')
plt.xlabel('Année')
plt.ylabel('Masse')
plt.scatter(x=planets.year, y=planets.mass)
plt.show()
```

### Diagramme circulaire

> Un graphique circulaire ou graphique en secteurs, aussi appelé camembert en France, est un type de graphique utilisé en statistiques. Il permet de représenter un petit nombre de valeurs par des angles proportionnels à la fréquence de ces valeurs. (wikipedia)

```python
years = planets.year.unique().tolist()
years = sorted(years)
```

```python
vals = planets.year.value_counts().to_list()[:-5]
pie_labels =[str(year) for year in years[:-5]]

explode = [0.1 for i in range(len(years[:-5]))]
```

```python
plt.pie(x=vals, 
        explode=explode, 
        shadow=True, 
        labels=pie_labels, 
        radius=2, 
        autopct=f'%.1f%%')

plt.show()

# on s'arrête à 2009 car les autres valeurs sont beaucoup trop petites
```

### Grille par paire de variables

> Assez similaire à la courbe par paire de variables, mais celle-ci donne la possibilité "mapper" la courbe via une fonction de matplotlib 

```python
#on crée d'abord la grille
grid = sns.PairGrid(data=planets, hue="year")
# puis on applique la fonction matplotlib voulue
grid = grid.map_diag(plt.hist)
```

```python
# elle est un peu vide cette grille...
# et si on remplissait les cases vides avec des nuages de points?

#on crée d'abord la grille
grid = sns.PairGrid(data=planets, hue="year")
# puis on applique la fonction matplotlib voulue
grid = grid.map_diag(plt.hist)

# on rajoute maintenant les points 
grid = grid.map_offdiag(plt.scatter)
```

### Un dernier exemple avec un graphique 3D

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# matplotlib version
fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig)

xs = planets['distance']
ys = planets['mass']
zs = planets['year']
ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')

ax.set_xlabel('distance')
ax.set_ylabel('Mass')
ax.set_zlabel('Year')

plt.show()
```
